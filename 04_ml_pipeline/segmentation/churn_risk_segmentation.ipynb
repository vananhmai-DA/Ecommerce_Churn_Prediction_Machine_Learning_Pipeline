The prediction output was merged with the Gold customer table to analyze churn risk segments.

High Risk customers show several clear patterns. More than half of them have submitted complaints, suggesting that complaint handling is a key retention lever.

High Risk customers also have very low tenure. The median tenure of this group is only 1, which indicates that many churn-risk customers are still in the early stage of their customer lifecycle. This suggests that onboarding and early engagement campaigns are important.

In terms of preferred order category, High Risk customers are concentrated in mobile-related categories, especially mobile phone and mobile. This may suggest that these categories are more one-off purchase driven and may require different retention strategies compared with repeat-driven categories such as grocery.

The High Risk customer segment was saved to `05_outputs/segments/high_risk_customers.csv` for further business recommendation.

## Segmentation Summary

The prediction output was merged with the Gold customer table to analyze churn risk segments.

The model classified 853 customers as High Risk, representing around 15.15% of the total customer base. Most customers were classified as Low Risk, while only a small share fell into the Medium Risk group.

High Risk customers show several clear patterns. More than half of them have submitted complaints, suggesting that complaint handling is a key retention lever.

High Risk customers also have very low tenure. The average tenure of the High Risk group is around 2.63, compared with around 11.07 for Low Risk customers. This suggests that many churn-risk customers are still in the early stage of their customer lifecycle. Onboarding and early engagement are therefore critical.

High Risk customers are concentrated in mobile-related categories, especially mobile phone and mobile. This may suggest that these categories are more one-off purchase driven and may require different retention strategies compared with repeat-driven categories such as grocery.

High Risk customers also have lower average cashback amounts than Low Risk customers. This may suggest that incentives or personalized retention offers could help improve engagement, especially for customers with low tenure and complaint history.

Some variables should be interpreted carefully. For example, satisfaction score is higher on average among High Risk customers, which does not follow the usual expectation. Therefore, satisfaction-related variables should be further validated before making strong business conclusions.

The High Risk customer segment was saved to `05_outputs/segments/high_risk_customers.csv` for further business recommendation.
